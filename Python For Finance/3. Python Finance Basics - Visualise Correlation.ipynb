```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

### load data and get correlations to save it as CSV file

```python
df = pd.read_csv('sp500_joined_closes.csv')
```

```python
df = df.set_index("Date")
df = df.fillna(0)
```

```python
df.tail()
```

```python
corr = df.corr()
corr.to_csv("sp500corr.csv")
```

### load correlation data and visualize it with heat map plot in seaborn

```python
sns.heatmap(corr)
```

### visualize the correlation between Amzon and other stocks

```python
cols = corr.nlargest(10, "AMZN")['AMZN'].index
```

```python
#scatterplot pair plot
sns.set()
sns.pairplot(df[cols])
```
