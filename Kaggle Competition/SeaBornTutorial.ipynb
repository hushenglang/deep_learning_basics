### Histograms

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
```

```python
for col in ["x","y"]:
    plt.hist(data[col], normed=True, alpha=0.5)
```

### Kernel Desity Estimator

```python
# density estimator, similar to histo gram.b
for col in ["x","y"]:
    sns.kdeplot(data[col], shade=True)
```

```python
sns.kdeplot(data)
```

### Combination of histo gram and density estimatorb

```python
# combination of histo gram and density estimatorb
for col in ["x","y"]:
    sns.distplot(data[col])
```

### Joint plot

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex');
```

### !Pairplots - Very Useful
When you generalize joint plots to data sets of larger dimensions, you end up with pair plots. This is very useful for exploring correlations between multi-dimensional data, when youd like to plot all pairs of values against each other.

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', size=2.5);
```

### Faceted Histograms

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
```

```python
tips.head()
```

```python
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

### Factor Plots

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill");
```

### Joint Distributions

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=tips, kind='hex')
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```
